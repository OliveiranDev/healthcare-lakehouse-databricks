# Lakehouse Setup - Camada Landing
Este notebook prepara a insfraestrutura inicial do projeto **Healthcare Lakehouse** no Databricks.

O objetivo é criar os componentes necessários para ingestão de dados utilizando a arquitetura **Medallion**.

## Componentes criados
- Catálogo do projeto
- Schemas das camadas do Lakehouse
- Volumes para armazenamento de arquivos brutos

## Arquitetura de dados
Fluxo da pipeline:

Landing > Bronze > Silver > Gold

|  Camada  |  Objetivo  |
|------|------|
|  landing  |  recepção de arquivos brutos  |
|  bronze  |  ingestão inicial Delta  |
|  silver  | limpeza e padronização  |
|  gold  |  datasets analíticos  |

---

## Organização dos volumes
A camada landing possui três volumes principais:
|  Volume  |  Função  |
|------|------|
|  raw  |  arquivos recebidos da fonte  |
|  stage  |  arquivos convertidos ou preparados  |
| archive  |  histórico de arquivos já processados  |


## Cria catálogo

```python
%sql
CREATE CATALOG IF NOT EXISTS healthcare_dev
COMMENT 'Catálogo principal do projeto Healthcare Lakehouse';
```

## Define contexto

```python
%sql
USE CATALOG healthcare_dev;
```

## Cria schemas
cada camada da arquitetura Medallion será representada por um schema.
|  Schema  |  Camada  |
|------|------|
|  landing  |  arquivos brutos  |
|  bronze  | dados ingeridos  |
|  silver  | dados tratados  |
|  gold  | dados analíticos  |

```python
%sql
CREATE SCHEMA IF NOT EXISTS healthcare_dev.landing
COMMENT 'Landing zone - recepção de dados brutos ';

CREATE SCHEMA IF NOT EXISTS healthcare_dev.bronze
COMMENT 'Camada bronze - dados brutos preservados em Delta';

CREATE SCHEMA IF NOT EXISTS healthcare_dev.silver
COMMENT 'Camada silver - dados limpos, validados e padronizados';

CREATE SCHEMA IF NOT EXISTS healthcare_dev.gold
COMMENT 'Camada gold - dados agregados e preparados para consumo';
```

## Cria volume
Volume são objetivos do Unity Catalog utilizados para armazenar arquivos não tabulares.

Neste projeto a camada landing foi organizada em três volumes:

**raw** - arquivos recebidos da origem
**stage** - arquivos preparados para ingestão
**archive** - histórico de arquivos já processados

```python
%sql
CREATE VOLUME IF NOT EXISTS healthcare_dev.landing.raw
COMMENT 'Arquivos brutos recebidos das fontes de dados';

CREATE VOLUME IF NOT EXISTS healthcare_dev.landing.stage
COMMENT 'Arquivos preparados para ingestão na camada bronze';

CREATE VOLUME IF NOT EXISTS healthcare_dev.landing.archive
COMMENT 'Arquivos históricos ou já processados';
```
