Camada Gold - Setup

Este notebook prepara os schemas e estruturas operacionais da camada gold.

##Objetivo

- Cria schema gold e gold_ops
- cria checkpoint do gold para incrementalidade por ingestion_ts(fontes silver)
- cria/garante quality_reports_history, onde será registrado volume, última competência, último build e status do snapshot.

##Estratégia

- SQL-only
- Checkpoint em gold_ops.pipeline_checkpoint

##Contexto

```sql
USE CATALOG healthcare_dev;
```

##Cria schemas

```sql
CREATE SCHEMA IF NOT EXISTS healthcare_dev.gold;
CREATE SCHEMA IF NOT EXISTS healthcare_dev.gold_ops;
```

##Checkpoint gold

```sql
CREATE TABLE IF NOT EXISTS healthcare_dev.gold_ops.pipeline_checkpoint (
  table_name STRING,
  last_ingestion_ts TIMESTAMP,
  last_run_ts TIMESTAMP,
  status STRING
)
USING DELTA;
```

##Cria tabela quality_reports_history

```sql
CREATE TABLE IF NOT EXISTS healthcare_dev.gold_ops.quality_report_history (
  report_ts TIMESTAMP,
  table_name STRING,

  row_count BIGINT,
  max_competencia STRING,
  max_gold_build_ts TIMESTAMP,

  status STRING,
  notes STRING
)
USING DELTA;
```
